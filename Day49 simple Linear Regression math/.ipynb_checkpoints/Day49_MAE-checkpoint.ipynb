```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(data = df , x = 'cgpa',y='package')
```

```python
import seaborn as sns
sns.scatterplot(data = df , x='cgpa',y='package')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
#lr.predict(X_test.iloc[2].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
y_test.values
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
print("R2_score",r2_score(y_test,y_pred))
r2 = r2_score(y_test,y_pred)
```

```python
r2
```

```python
X_test.shape
```

```python
# adjusted R2 Score
```

```python
ar2 = 1 - ((1 - r2) * 39) / (39 - 1)
```

```python
ar2
```

```python
def adjusted_r_squared(r_squared, n, p):
    return 1 - ((1 - r_squared) * (n - 1)) / (n - p - 1)

# Example values
r_squared = r2  # Replace this with your actual R^2 value
n = 40         # Total number of observations
p = 1            # Number of predictors

adj_r_squared = adjusted_r_squared(r_squared, n, p)
print("Adjusted R-squared:", adj_r_squared)
```
