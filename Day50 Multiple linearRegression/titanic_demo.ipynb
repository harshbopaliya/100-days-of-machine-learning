```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score
```

```python
age_imputer = SimpleImputer(strategy='median')
train_data['Age'] = age_imputer.fit_transform(train_data[['Age']])
test_data['Age'] = age_imputer.transform(test_data[['Age']])
```

```python
train_data.isnull().sum()
```

```python
embarked_imputer = SimpleImputer(strategy='most_frequent')
train_data['Embarked'] = embarked_imputer.fit_transform(train_data[['Embarked']]).ravel()
```

```python
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
X_train = train_data.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin', 'Fare'])
y_train = train_data['Fare']

X_test = test_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train_split, y_train_split)
```

```python
y_pred = model.predict(X_val_split)
```

```python
mse = mean_squared_error(y_val_split, y_pred)
r2 = r2_score(y_val_split, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
